```python
df10 = pd.read_csv('Fifa19.csv',encoding='unicode_escape')
```

```python
df10.drop(['ID','Name','Photo','Nationality','Flag','Club','Club Logo','Preferred Foot','Unnamed: 0','Work Rate','Body Type','Real Face','Joined','Loaned From'], axis = 1, inplace=True)
```

```python
df10.dtypes
```

```python
df10 = df10.fillna(0)
```

```python
a = []
b = []
c = []
for i in df10.Wage:
    a.append(str(str(i).replace('\x80','')))
for j in df10.Value:
    b.append(str(str(j).replace('\x80','')))
for k in df10['Release Clause']:
    c.append(str(str(k).replace('\x80','')))    
```

```python
df10['Wage'] = a
df10['Value'] = b
df10['Release Clause'] = c
```

```python
d = []
for l in df10['Weight']:
    d.append(str(str(l).replace('lbs','')))
df10['Weight'] = d
```

```python
e = []
f = []
for m in df10['Height']:
    m1 = int(str(m)[0])
    m2 = str(m)[2:]
    e.append(m1)
    f.append(m2)
```

```python
df10['Height Feet'] = e
df10['Height Inch'] = f
df10.drop('Height', inplace = True, axis = 1)
```

```python
df10['Height Inch'] = pd.to_numeric(df10['Height Inch'])
```

```python
df10['Height'] = (12*df10['Height Feet']) + df10['Height Inch']
```

```python
df10['Height'] = df10['Height'] * 2.54
```

```python
df10.drop(['Height Feet','Height Inch'], axis = 1, inplace = True)
```

```python
def final(x2):
    for x1 in x2:
        x2 = []
        for n in df10[x1]:
            n1 = int(str(n)[:2])
            n2 = int(str(n)[-1])
            x2.append(n1+n2)
        df10[x1] = x2
```

```python
def MKC(arg):
    z = df10[arg]
    g = []
    for o in z:
        if str(o)[-1] == 'K':
            g.append(float(str(o)[:-1])*1000)
        elif str(o)[-1] == 'M':
            g.append(float(str(o)[:-1])*1000000)
        else:
            g.append(o)
    df10[arg] = g
```

# Bagging and Boosting

```python
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
#Bagging
bag = BaggingRegressor(base_estimator=model, n_estimators=200)
bag_model = bag.fit(xtrain, ytrain)
bag_ypred = bag_model.predict(xtest)
r2_score(ytest, bag_ypred)
```

```python
#Adaboost
aboost = AdaBoostRegressor(base_estimator=model, n_estimators=200, learning_rate=0.0001, loss='exponential')
aboost_model = aboost.fit(xtrain, ytrain)
aboost_ypred = aboost_model.predict(xtest)
r2_score(ytest, aboost_ypred)
```

```python
#Gradient Boosting
gboost = GradientBoostingRegressor(loss='huber', n_estimators=200, learning_rate=0.001)
gb_model = gboost.fit(xtrain, ytrain)
```

```python
gb_ypred = gb_model.predict(xtest)
r2_score(ytest, gb_ypred)
```
