```python
scale = StandardScaler()
df_num = df6[num]
scaled_df_num = scale.fit_transform(df_num)
```

```python
df_num = pd.DataFrame(scaled_df_num, columns=num)
df_label = df6[labels]
```

```python
df1 = pd.concat((df_num,df_label), axis = 1)
```

```python
features = df1.drop('team_position1', axis = 1)
target = df1.team_position1
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size = 0.2)
```

```python
lr = LogisticRegression(max_iter=200)
knn = KNeighborsClassifier(n_neighbors=5)
svc = SVC(kernel='rbf')
rfc = RandomForestClassifier(n_estimators=200,criterion='gini')
gauss = GaussianNB()
bern = BernoulliNB()
cat = CategoricalNB()
xgbc = XGBClassifier()
# mult = MultinomialNB()
alg = [lr,knn,svc,rfc,gauss,bern,xgbc]
```

```python
for i in alg:
    i.fit(xtrain, ytrain)
    ypred = i.predict(xtest)
    print('{} : '.format(i), accuracy_score(ytest, ypred))
```

```python
# With Boosting
alg2 = alg.remove(['knn'])
for i in alg2:
    i.fit(xtrain, ytrain)
    boost = AdaBoostClassifier(base_estimator=i,n_estimators=200).fit(xtrain, ytrain)
    ypred = boost.predict(xtest)
    print('{} : '.format(i), accuracy_score(ytest, ypred))
```

```python
# With Bagging
for i in alg:
    i.fit(xtrain, ytrain)
    bag = BaggingClassifier(base_estimator=i,n_estimators=200).fit(xtrain, ytrain)
    ypred = bag.predict(xtest)
    print('{} : '.format(i), accuracy_score(ytest, ypred))
```
