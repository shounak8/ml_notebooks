```python
features = df.drop('team_position1', axis = 1)
target = df.team_position1
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size = 0.2)
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier(n_neighbors=5)
svc = SVC(kernel='rbf')
rfc = RandomForestClassifier(n_estimators=200,criterion='gini')
gauss = GaussianNB()
bern = BernoulliNB()
# cat = CategoricalNB()
mult = MultinomialNB()
alg = (lr,knn,svc,rfc,gauss,bern,mult)
```

```python
lr.fit(xtrain, ytrain)
ypred = lr.predict(xtest)
print(i, accuracy_score(ytest, ypred))
```

```python
for i in alg:
    i.fit(xtrain, ytrain)
    ypred = i.predict(xtest)
    print(i, accuracy_score(ytest, ypred))
```
