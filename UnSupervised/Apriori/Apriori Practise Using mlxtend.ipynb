```python
df = pd.read_csv('OnlineRetail.csv', encoding='unicode_escape')
```

```python
df['Country'].value_counts()
```

```python
# Analyzing for Germany
df_Ger = df[df['Country'] == 'Germany']
```

```python
df_Ger
```

```python
df_Ger['Quantity'].sort_values(ascending=True)
```

```python
# Removing the negative Quantity rows
df_Ger = df_Ger[df_Ger['Quantity'] > 0]
```

```python
df_Ger = df_Ger.fillna(0)
```

```python
df_Ger1 = df_Ger[['InvoiceNo','Description']]
```

```python
df_Ger1['Qty'] = np.nan
```

```python
df_Ger1 = df_Ger1.fillna(1)
```

```python
df_Ger1 = df_Ger1.groupby(['InvoiceNo','Description'])['Qty'].sum()
```

```python
df_Ger1 = df_Ger1.unstack().fillna(0)
```

```python
df_Ger1
```

```python
df_Ger1 = df_Ger1.applymap(lambda x: 0 if x == 0 else 1)
```

```python
support = apriori(df_Ger1, min_support= 0.05, use_colnames=True)
```

```python
rules = association_rules(support, metric='lift', min_threshold=1)
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
