```python
import os
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
os.chdir('C:\\Users\\shoun\\Coding\\Datasets')
```

```python
df3 = pd.read_csv('rawDataset.csv',sep='delimiter', header=None)
```

```python
df3
```

```python
a = []
for i in df3[0]:
    a1 = str(i).split(',')[2:]
    a.append(a1)
```

```python
a
```

```python
'''removing blank cells'''
for i in a:
    while ' ' in i:
        i.remove(' ')
```

```python
a
```

```python
encode = TransactionEncoder()
encode.fit_transform(a)
```

```python
items = encode.columns_
```

```python
items[0] = 'delete'
```

```python
items
```

```python
df2 = pd.DataFrame(encode.fit_transform(a), columns=items)
```

```python
df2.drop('delete', inplace=True, axis = 1)
```

```python
df2
```

```python
model = apriori(df2,min_support=0.001, max_len=3, use_colnames=True)
rule = association_rules(model, metric='lift', min_threshold=1)
rule.sort_values(by = 'confidence', ascending= False)
```
