```python
import os
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
os.chdir('C:\\Users\\shoun\\Coding\\Datasets')
df5 = pd.read_csv('rawDataset.csv',sep='delimiter', header=None)
```

```python
df5[0] = df5[0].str.strip()
```

```python
df4 = df5[0].str.split(',', expand=True)
```

```python
df4 = df4.replace(r'^\s*$', np.nan, regex=True)
```

```python
df3 = df4.iloc[:,2:]
```

```python
df3 = df3.fillna(0)
```

```python
df3
```

```python
l = []
for i in range(0,len(df3)):
    l.append([str(df3.values[i,j])for j in range(0,9)])
```

```python
l
```

```python
for i in range(0,len(l)):
    while '0' in l[i]:
        l[i].remove('0')
```

```python
l
```

```python
model = apriori(l, min_support = 0.01, min_confidence = 0.02, min_lift = 1, min_length=1)
rule = list(model)
```

```python
for b in rule:
    pair = b[0]
    bb = [k for k in pair]
    print(bb)
```

```python
result = []
for item in rule:
    pair = item[0]
    items = [x for x in pair]
    
    value0 = str(items[0])
    value1 = str(item[2][0][1])
    value2 = str(item[1])[:7]
    value3 = str(item[2][0][2])[:7]
    value4 = str(item[2][0][3])[:7]
    
    rows = (value0,value1,value2,value3,value4)
    result.append(rows)
    
    label = ['antecedant','consequant','support','confidence','lift']
    
    rule1 = pd.DataFrame.from_records(result, columns=label)
    
    print(rule1)
```

```python
z = pd.DataFrame(rule1)
```
