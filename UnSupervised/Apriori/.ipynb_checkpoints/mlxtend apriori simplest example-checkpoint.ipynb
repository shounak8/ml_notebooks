```python
import os
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df6 = pd.read_csv('groceries.csv', sep = 'delimited', header = None)
```

```python
type(df6[0])
```

```python
a = []
for i in range(0,len(df6[0])):
    a.append(str(df6[0][i]))
```

```python
a
```

```python
b = []
for i in a:
    b.append(str(i).split(','))
```

```python
b
```

```python
encode = TransactionEncoder()
encode.fit_transform(b)
```

```python
df5 = pd.DataFrame(encode.fit_transform(b), columns=encode.columns_)
```

```python
df5
```

```python
model = apriori(df5, min_support= 0.05, max_len=2,use_colnames = True)
```

```python
rule = association_rules(model,metric='lift', min_threshold=1)
```

```python
rule.sort_values(by = 'lift', ascending=False)
```
