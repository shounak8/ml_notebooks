```python
# Form a function for above steps:
# 1. allot url to variable 'a'
# 2. read the url using urllib.request.urlopen(a).read().decode('utf8','ignore')
# 3. pass it through BeautifulSoup to parse the data
# 4. find all articles you are looking for (using 'find_all') and extract the text

def getTextWaPo(url):
    page = urllib.request.urlopen(url).read().decode('utf','ignore')
    soup = BeautifulSoup(page,'lxml')
    text = ' '.join(map(lambda x:x.text, soup.find_all('article')))
    text = text.replace('?',' ')
    return text.replace('\xa0',' ')
```

```python
text = getTextWaPo(link)
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
stopwords_punctuation = set(stopwords.words('english')+ list(punctuation))
```

```python
# removing stop words & punctuation
word_sent = [word for word in word_sent if word not in stopwords_punctuation]
```

```python
word_sent
```

```python
aa = ['”','“','’']
```

```python
for i in aa:
    while i in word_sent:
        word_sent.remove(i)
```

```python
word_sent
```

```python
# Create frequency distribution of words
from nltk import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
# Calculate significance score for article
from collections import defaultdict
# defaultdict will add a key to the dictionary in case there is no key in the pre-existing dictionary
```

```python
# create blank dictionary called ranking
ranking = defaultdict(int)

# we will calculate the significance of the sentences by checking the words in each sentence and 
# adding the frequency of those words the we calculated in freq

for i,sent in enumerate(sents):
    for j in word_tokenize(sent.lower()):
        if j in freq:
            ranking[i] += freq[j]
ranking
```

```python
# this ranking has given us the significance scores of all sentences
```

```python
# if you want the summary to be based on 5 sentences, pick the top 5 sentences based on the significance
from heapq import nlargest

sents_idx = nlargest(5, ranking, key = ranking.get)
sents_idx
# this gives the index number of sentences giving the max significance
```

```python
# above sents_idx is sorted as per significance score. Sort the sentences as per the article sequence
sorted_sent = [sents[i] for i in sorted(sents_idx)]
sorted_sent
```

```python
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n<= len(sents)
    word_sent = word_tokenize(text.lower())
    
    word_sent = [word for word in word_sent if word not in set(stopwords.words('english') + list(punctuation))]
    freq = FreqDist(word_sent)
    
    ranking = defaultdict(int)
    
    for i, sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
                
    sents_idx = nlargest(n, ranking, key = ranking.get)
    
    return [sents[j] for j in sorted(sents_idx)]
```
